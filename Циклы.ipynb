```python
var = 1          # создаем переменную, равную 1
while var <= 13: # прописываем цикл с условием - выполнять до тех пор, пока переменная меньше или равна 13
    print(var)   # выводим значение переменной
    var += 1     # увеличиваем переменную на 1
```

```python
for var in 'Python':
    if var == 'h':
        continue
    print(var, end='')
```

```python
for var in 'Python':
    if var == 'h':
        break
    print(var)
```

```python
for var in 'Python':
    if var == 'a':
        break 
else:
    print('Символа a нет в слове Python')
```

```python
zero = 0
chisla = [700, 8, 89, 20, 13]
for i in chisla:
    if i == 0:
        zero += 1
print(zero)
```
